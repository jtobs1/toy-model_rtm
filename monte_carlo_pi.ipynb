Monte Carlo to calculate $\pi$

```python
import numpy as np
import matplotlib.pyplot as plt

# Create the Cartesian Domain
n_points = 10000
x = np.linspace(0, 1, n_points)
y = np.linspace(0, 1, n_points)
X, Y = np.meshgrid(x, y)

# generate a ton of random point locations from 0 to 1
x_point_loc = np.random.rand(len(x))
y_point_loc = np.random.rand(len(y))
xy_loc = np.array([x_point_loc, y_point_loc])
```

```python
# Create the "Circle Condition"
cond = np.zeros_like(xy_loc)
for i in range(len(x)):
    if (np.sqrt(xy_loc[0, i]**2 + xy_loc[1, i]**2) <= 1):
        cond[0, i] = True
        cond[1, i] = True
    elif (np.sqrt(xy_loc[0, i]**2 + xy_loc[1, i]**2) > 1):
        cond[0, i] = False
        cond[1, i] = False
    else:
        print('error')
```

```python
total_points = cond.shape[0] * cond.shape[1]
true_points = cond.sum()
# the "known area" of the surface of the circle (quarter)
known_area = 4

pi_estimate = (known_area * true_points) / total_points

print(pi_estimate)
```

```python
fig, ax = plt.subplots(figsize=(6, 5))
plt.title(f'Pi Estimate with {n_points} Points: {pi_estimate}')
plt.scatter(xy_loc[0,][cond[0]==1], xy_loc[1,][cond[1]==1], s=1, c='r')
plt.scatter(xy_loc[0,][cond[0]!=1], xy_loc[1,][cond[1]!=1], s=1, c='b')
plt.show()
plt.close()
```
